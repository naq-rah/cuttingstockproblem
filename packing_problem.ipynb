```
  B=20
  w = [9,8,7,6]
  q = [511,301,263,383]
```

```

# B =  12             #stock size 
# w = [2,3,4,5,6]     # product variety (different sizes ) 
# q = [1,2,3,2,4]     #order quantity 
s=[]                #list containg sizes. 
for j in range(len(w)):
    for i in range(q[j]):
        s.append(w[j])
print(s)
```

```
def FFD(s, B):
    remain = [B]
    sol = [[]]
    for item in sorted(s, reverse=True):
        for j,free in enumerate(remain):
            if free >= item:
                remain[j] -= item
                sol[j].append(item)
                break
        else:
            sol.append([item])
            remain.append(B-item)
    return sol
```

```
print(FFD(s,B))
print(len(FFD(s,B)))   # U = len(FFD)
```

```
row_indices=np.arange(len(s))
```

```
col_indices=np.arange(len(FFD(s,B)))
```

```
#declare the decision variables in the model
model.x = Var(row_indices,col_indices, domain=Binary)
```

```
model.y= Var(col_indices, domain=Binary)
```

```
model.objective=Objective(expr=summation(model.y),sense=minimize)
```

```
model.constraints=ConstraintList()
```

```
# print(row_indices)
# print(col_indices)
```

```
#constraintr1
for i in row_indices:
  model.constraints.add(expr=sum(model.x[i,j] for j in col_indices) == 1)
#constraint2
for j in col_indices:
  model.constraints.add(expr=sum(s[i]*model.x[i,j] for i in row_indices) <= B*model.y[j])
#constraint3
for j in col_indices:
    for i in row_indices:
       model.constraints.add(expr=(model.x[i,j] <= model.y[j]))
```
